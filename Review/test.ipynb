```python
%matplotlib inline
import matplotlib.pyplot as plt
# import pandas as pd
import numpy as np

x = np.array([1,2,3,4])
rand1 = np.random.random(10)
rand2 = np.random.random(10)
randn = np.random.randn(1000)
nut_prices = {"Almond": 10,
              "Peanut": 8,
              "Cashew": 12}

# Method 1: using tuples
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,10))
# plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(rand1, rand2);
ax3.bar(nut_prices.keys(),nut_prices.values());
ax4.hist(randn);

# Customize subplots
ax1.set(title="Line SubPlot", xlabel="x", ylabel="x/2");
ax2.set(title="Scatter SubPlot", xlabel="x", ylabel="y");
ax3.set(title="Bar SubPlot", xlabel="Nut", ylabel="Price ($)");
ax4.set(title="Histogram SubPlot", xlabel="Value", ylabel="Frequency");
```
