#### Pyplot vs. Object Oriented (OO) Interface

* While it is easy to quickly generate plots with the matplotlib.pyplot module, the use of <span style="color:hotpink">object-oriented approach is recommended</span> as it gives more <span style="color:hotpink">control and customization</span> of your plots.
* The main idea behind using the more formal object-oriented method is to <span style="color:hotpink">create figure objects and then just call methods or attributes off of that object</span>. 
* In object-oriented interface, Pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects.
* The <span style="color:hotpink">pyplot approach doesn't really scale when we are required to make multiple plots </span>or when we have to make intricate plots that require a lot of customisation

Pyplot Interface:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

plt.figure()
plt.plot(x,y)
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["legend"])
plt.show()
```

OO Interface (3 methods)

```python
# 1st method
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig = plt.figure() # creates a figure instance (empty canvas)
ax = fig.add_subplot() # adds some axes
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
fig.legend(["legend"])
plt.show()
```

```python
# 2nd method
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig = plt.figure() # creates a figure instance (empty canvas)
ax = fig.add_axes([1,1,1,1]) # [left, bottom, width, height]
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
fig.legend(["legend"])
plt.show()
```

```python
# 3rd method (recommended)
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig, ax = plt.subplots() # mult subplots in same fig
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
fig.legend(["legend"])
plt.show()
```

Example of multiple plots...

```python
# Pyplot Interface
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

plt.figure()
plt.plot(x,y)
plt.plot(x,np.square(x))
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["x", "x^2"])
plt.show()
```

```python
# Object Oriented Interface
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig, ax1 = plt.subplots()
ax1.set_title("Title")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.plot(x,y,"blue")

ax2 = ax1.twinx() # create another y-axis sharing a common x-axis
ax2.set_xlabel("x-axis")
ax2.set_ylabel("y-axis")
ax2.plot(x,np.square(x),"orange")

fig.legend(["x","x^2"])
plt.show()
```

Example Workflow

```python
# 0. import tools
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prep data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(3,3)) # (width, height)

# 3. Plot Data
ax.plot(x,y);

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & share
fig.savefig("./sample-plot2.png")
```
