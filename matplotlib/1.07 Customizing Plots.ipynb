## 1.07 Customizing Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# First we need to import our data
car_sales = pd.read_csv("car-sales.csv")

# Let's see what the data looks like and check the types
car_sales, car_sales.dtypes
```

```python
# In order to plot our data, we need to convert price to a numeric type

# Remove the dollar signs and commas from the price column
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")

# Remove extra zeros
car_sales["Price"] = car_sales["Price"].str[:-2]

# Convert to integer type
car_sales["Price"] = car_sales["Price"].astype(int)

# Check data type
type(car_sales["Price"][0]), car_sales["Price"].dtype, car_sales.dtypes
```

```python
# Import heart disease data
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# limit data to patients over 50
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

### Changing the title, legend and axes

```python
# Create some data
x = np.random.rand(10, 4)
x
```

### Changing the cmap

`cmap` - colormap; the color map is a mapping from data values to colors.

[See documentation for more options](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
# No cmap change
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50['age'], 
                     over_50['chol'], 
                     c=over_50['target'])
ax.set(title='Heart Disease and Cholesterol Levels', 
       xlabel='Age', 
       ylabel='Cholesterol')
ax.axhline(y=over_50['chol'].mean(), 
           c='b', 
           linestyle='--', 
           label='Average');
ax.legend(*scatter.legend_elements(), title='Target');
```

```python
# Change cmap and horizontal line to be a different color
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50['age'], 
                     over_50['chol'], 
                     c=over_50['target'],
                     cmap='winter') # color map - changes the color scheme.
ax.set(title='Heart Disease and Cholesterol Levels', 
       xlabel='Age', 
       ylabel='Cholesterol')
ax.axhline(y=over_50['chol'].mean(), 
           c='r', 
           linestyle='--', 
           label='Average');
ax.legend(*scatter.legend_elements(), title='Target');
```

### Changing the limits

```python
# Before setting axis limits
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 10))

# For the first plot (ax0):
scatter = ax0.scatter(over_50['age'], 
                      over_50['chol'], 
                      c=over_50['target'], 
                      cmap='winter')
ax0.set(title='Heart Disease and Cholesterol Levels', 
        ylabel='Cholesterol')
ax0.axhline(y=over_50['chol'].mean(), 
            color='r', 
            linestyle='--', 
            label='Average');
ax0.legend(*scatter.legend_elements(), 
           title='Target')

# For the second plot (ax1):
scatter = ax1.scatter(over_50['age'], 
                      over_50['thalach'], 
                      c=over_50['target'], 
                      cmap='winter')
ax1.set(title='Heart Disease and Max Heart Rate Levels', 
        xlabel='Age', 
        ylabel='Max Heart Rate')
ax1.axhline(y=over_50['thalach'].mean(), 
            color='r', 
            linestyle='--', 
            label='Average');
ax1.legend(*scatter.legend_elements(), 
           title='Target')

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
# recreate the plots above adding x & y limitations
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 10))

# For the first plot (ax0):
scatter = ax0.scatter(over_50['age'], 
                      over_50['chol'], 
                      c=over_50['target'], 
                      cmap='winter')
ax0.set(title='Heart Disease and Cholesterol Levels', 
        ylabel='Cholesterol')
ax0.set_xlim(50, 80) # set x-axis limits
ax0.axhline(y=over_50['chol'].mean(), 
            color='r', 
            linestyle='--', 
            label='Average');
ax0.legend(*scatter.legend_elements(), 
           title='Target')

# For the second plot (ax1):
scatter = ax1.scatter(over_50['age'], 
                      over_50['thalach'], 
                      c=over_50['target'], 
                      cmap='winter')
ax1.set(title='Heart Disease and Max Heart Rate Levels', 
        xlabel='Age', 
        ylabel='Max Heart Rate')
ax1.set_xlim(50, 80) # set x-axis limits
ax1.set_ylim(60, 200) # set y-axis limits
ax1.axhline(y=over_50['thalach'].mean(), 
            color='r', 
            linestyle='--', 
            label='Average');
ax1.legend(*scatter.legend_elements(), 
           title='Target')

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
