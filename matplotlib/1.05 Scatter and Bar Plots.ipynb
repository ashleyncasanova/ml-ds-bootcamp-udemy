## 1.5 Scatter and Bar Plots

**Plot Types:**
* `Line plot`
* `Scatter plot`
* `Bar plot`
* `Histogram`
* `Subplots`

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

> Note: Matplotlib visualizations are built off Numpy arrays.

### Line Plot

```python
# Create some data
x = np.linspace(0, 10, 100)

# np.linspace returns evenly spaced numbers over a specified interval.
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

>Note: The default plot in matplotlib, when you call `plot()`, is a line plot.

### Scatter Plot

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

>Note: Must reset figure and axis if you are in a new cell.

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar Plot

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Prices", ylabel="Price($)");
```

```python
# we can create a horizontal bar plot using barh
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Dan's Nut Butter Prices", xlabel="Price($)");
```

### Histogram

```python
# Make some data from a normal distribution and plot it

x = np.random.randn(1000) 
# np.random.randn() returns a random number with a normal distribution.

fig, ax = plt.subplots()
ax.hist(x)

ax.set(title="Histogram of a Normal Distribution", xlabel="Value", ylabel="Frequency");
```

```python
# Make some data from a random distribution and plot it

x = np.random.random(1000)
# np.random.random() returns a random number with uniform distribution.

fig, ax = plt.subplots()
ax.hist(x);

ax.set(title="Histogram of a Random Distribution", xlabel="Value", ylabel="Frequency");
```

### Subplots
* Multiple plots in one figure

```python
# Method 1: Create multiple subplots using tuples
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 10))

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));

# Customize the subplots
ax1.set(title="Line Subplot", xlabel="x", ylabel="x/2");
ax2.set(title="Scatter Subplot", xlabel="x", ylabel="y");
ax3.set(title="Bar Subplot", xlabel="Nut Butter", ylabel="Price ($)");
ax4.set(title="Histogram Subplot", xlabel="Value", ylabel="Frequency");
```

```python
# Method 2: Create multiple subplots using the list-index method
fig, ax = plt.subplots(nrows=2, 
                       ncols=2, 
                       figsize=(10, 10))

# Use index coordinates to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));

# Customize the subplots
ax[0, 0].set(title="Line Subplot", xlabel="x", ylabel="x/2");
ax[0,1].set(title="Scatter Subplot", xlabel="x", ylabel="y");
ax[1,0].set(title="Bar Subplot", xlabel="Nut Butter", ylabel="Price ($)");
ax[1,1].set(title="Histogram Subplot", xlabel="Value", ylabel="Frequency");
```
